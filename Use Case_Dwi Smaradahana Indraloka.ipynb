## 3. Identify missing value. If any, how will you handle it?
* For numerical data we fill the missing value using the mean of the data
* For categorical data we fill the missing value using the mode of the data

```python
data.isna().sum()
```

```python
data["num-of-doors"].fillna(data["num-of-doors"].mode()[0], inplace = True)
```

```python
mean1 = data["bore"].mean()
data["bore"].fillna(mean1, inplace = True)
```

```python
mean2 = data["stroke"].mean()
data["stroke"].fillna(mean2, inplace = True)
```

```python
data["horsepower"].fillna(data["horsepower"].mode()[0], inplace = True)
```

```python
data["peak-rpm"].fillna(data["peak-rpm"].mode()[0], inplace = True)
```

```python
mean4 = data["price"].mean()
data["price"].fillna(mean4, inplace = True)
```

```python
data["normalized-losses"].fillna(data["normalized-losses"].mode()[0], inplace = True)
```

## 4. Perform visualization using at least 5 difference visualization technique

### a. Scatter Plot
* Engine Size VS Price
* Horse Power VS Price

```python
plt.scatter(data["engine-size"],data["price"])
plt.title("Engine Size vs Price")
plt.xlabel("Engine Size")
plt.ylabel("Price")
plt.show()

plt.scatter(data["horsepower"],data["price"])
plt.title("Horse Power vs Price")
plt.xlabel("Horse Power")
plt.ylabel("Price")
plt.show()
```

### b. Cross Table
* Car Brand VS Body Style
* Car Brand VS Cylinder

```python
table = pd.crosstab(data["make"], data["body-style"])
table.div(table.sum(1).astype(float), axis = 0).plot(kind="bar", stacked = True)

table = pd.crosstab(data["make"], data["cylinder"])
table.div(table.sum(1).astype(float), axis = 0).plot(kind="bar", stacked = True)
```

### c. Bar Plot
* Number of Cars Based on Body Style
* Number of Cars Based on Engine Type

```python
plt.hist(data["body-style"], bins = 9, rwidth = 0.9)
plt.title("Number of Cars Based on Body Style")
plt.xlabel("Body Style")
plt.ylabel("Number")
plt.show()

plt.hist(data["engine-type"], bins = 7, rwidth = 0.9)
plt.title("Number of Cars Based on Engine Type")
plt.xlabel("Engine Type")
plt.ylabel("Number")
plt.show()
```

### d. Box Plot
* Horse Power Distribution Data Based on Car Brand

```python
plt.figure(figsize = (15,8))
sns.set(style="whitegrid")
sns.boxplot(x = data["make"], y = data["horsepower"])
plt.xticks(rotation=45)
plt.title("Horse Power Distribution Data Based on Car Brand")
```

### e. Bar Chart
* 5 Highest Price of Car Brand with Two Doors, Gas Fuel Type and Four Cylinder
* 5 Lowest Price of Car Brand with Two Doors, Gas Fuel Type and Four Cylinder

```python
data_new = data[data["num-of-doors"]=="two"]
data_new1 = data_new[data_new["fuel-type"]=="gas"]
data_new2 = data_new1[data_new1["cylinder"]=="four"]
top_5_price = data_new2.sort_values("price", ascending = False).head(5)

top_5_price

x=range(5)
plt.bar(x,top_5_price["price"])
plt.xticks(x,top_5_price["make"])
plt.xlabel("Car Brand")
plt.ylabel("Price")
plt.title("5 Highest Price of Car Brand with Two Doors, Gas Fuel Type and Four Cylinder")
plt.show()

data_new = data[data["num-of-doors"]=="two"]
data_new1 = data_new[data_new["fuel-type"]=="gas"]
data_new2 = data_new1[data_new1["cylinder"]=="four"]
top_5_price = data_new2.sort_values("price").head(5)
top_5_price

x=range(5)
plt.bar(x,top_5_price["price"])
plt.xticks(x,top_5_price["make"])
plt.xlabel("Car Brand")
plt.ylabel("Price")
plt.title("5 Lowest Price of Car Brand with Two Doors, Gas Fuel Type and Four Cylinder")
plt.show()
```
